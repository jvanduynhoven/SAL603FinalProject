```python
# 1. Allow output from every line and import required packages
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd

# 2. Read in the NBA roster CSV files for the 2024–25 season and set player name as the index
lakers = pd.read_csv("lakers_2024_roster.csv", index_col="Player")
celtics = pd.read_csv("celtics_2024_roster.csv", index_col="Player")
warriors = pd.read_csv("warriors_2024_roster.csv", index_col="Player")
heat = pd.read_csv("heat_2024_roster.csv", index_col="Player")

# 3. Concatenate the four NBA team tables into one all-inclusive roster table
nba_rosters = pd.concat([lakers, celtics, warriors, heat])

# 4. Display only the first few rows of the concatenated table
display(nba_rosters.head())

# 5. Write the concatenated NBA roster table to a CSV file including the index
nba_rosters.to_csv("nba_2024_four_team_rosters.csv", index=True)

# 6. Verify the file was written correctly without displaying the table again
check_csv = pd.read_csv("nba_2024_four_team_rosters.csv", index_col="Player")
```
